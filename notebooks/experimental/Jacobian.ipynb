# 三角形元素幾何度量與 Jacobian 驗證

![本地圖片](./截圖%202026-03-27%2013.07.54.png)

在 Nodal DG 方法中，我們需要將標準三角形 $I$ 映射到實際的物理三角形 $D$ [cite: 52]。假設物理三角形的逆時針頂點為 $v^1, v^2, v^3$，我們可以透過仿射映射來建立轉換關係。

## 1. 座標映射 (Eq. 6.3)
物理座標 $\mathbf{x} = (x, y)$ 可以用標準坐標 $\mathbf{r} = (r, s)$ 表達 [cite: 83]：
$$x=-\frac{r+s}{2}v^1+\frac{r+1}{2}v^2+\frac{s+1}{2}v^3$$

## 2. 偏導數 (Eq. 6.4)
將上述映射對 $r$ 與 $s$ 進行偏微分，我們可以得到轉換度量 [cite: 93, 96]：
$$x_r = \frac{x_2 - x_1}{2}, \quad y_r = \frac{y_2 - y_1}{2}$$
$$x_s = \frac{x_3 - x_1}{2}, \quad y_s = \frac{y_3 - y_1}{2}$$

## 3. Jacobian 與逆度量項 (Eq. 6.5)
轉換的 Jacobian $J$ 以及鏈鎖律所需的逆度量項計算如下 [cite: 100, 101]：
$$J = x_r y_s - x_s y_r$$
$$r_x = \frac{y_s}{J}, \quad r_y = -\frac{x_s}{J}, \quad s_x = -\frac{y_r}{J}, \quad s_y = \frac{x_r}{J}$$

```python
import numpy as np

def compute_geometric_factors(v1, v2, v3):
    """
    根據給定的三個頂點座標，計算直邊三角形的 Jacobian 與幾何度量項。
    
    Parameters:
        v1, v2, v3 (array-like): shape (2,) 的陣列，代表 (x, y) 座標
        
    Returns:
        dict: 包含 J, rx, ry, sx, sy 等度量項的字典
    """
    x1, y1 = v1
    x2, y2 = v2
    x3, y3 = v3
    
    # 根據 Eq. 6.4 計算偏導數
    xr = (x2 - x1) / 2.0
    yr = (y2 - y1) / 2.0
    xs = (x3 - x1) / 2.0
    ys = (y3 - y1) / 2.0
    
    # 根據 Eq. 6.5 計算 Jacobian
    J = xr * ys - xs * yr
    
    if np.isclose(J, 0.0):
        raise ValueError("Jacobian為零！請檢查頂點是否共線，或是構成退化三角形。")
        
    # 計算逆矩陣的度量項
    rx = ys / J
    ry = -xs / J
    sx = -yr / J
    sy = xr / J
    
    return {
        "J": J,
        "rx": rx, "ry": ry,
        "sx": sx, "sy": sy,
        "xr": xr, "yr": yr,
        "xs": xs, "ys": ys
    }
```

```python
v1 = np.array([0.0, 0.0])
v2 = np.array([1.0, 0.0])
v3 = np.array([0.0, 1.0])

factors = compute_geometric_factors(v1, v2, v3)

print("=== 輸入頂點 ===")
print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"v3: {v3}\n")

print("=== 計算出的幾何度量項 ===")
print(f"Jacobian (J): {factors['J']}")
print(f"rx: {factors['rx']:>6.2f},  ry: {factors['ry']:>6.2f}")
print(f"sx: {factors['sx']:>6.2f},  sy: {factors['sy']:>6.2f}")
```
